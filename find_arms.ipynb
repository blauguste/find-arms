```python
from Bio import SeqIO
from Bio.SeqUtils import nt_search
import csv
```

#### Make list of promoters to search for

```python
promoters = ['NATCTTA', 'ANTCTTA', 'AANCTTA', 'AATNTTA', 'AATCNTA', 'AATCTNA', 'AATCTTN']
```

#### Read the MSA that we're going to be searching

```python
full_res = {}
```

```python
with open('hardticks.fasta', 'r') as infile:
    msa = SeqIO.parse(infile, 'fasta')
    for s in msa:
        promoter_res = []
        for p in promoters:
            results = nt_search(str(s.seq), p)
            promoter_res.append(results)
        full_res[s.id] = promoter_res
```

```python
full_res
```

```python
for k, v in full_res.items():
    print(k)
    for a in v:
        print(a[0])
        print(a[1:])
        for b in a[1:]:
            print(b)
```

```python
with open('arm_results.csv', 'w') as outfile:
    writer = csv.writer(outfile, delimiter = ',')
    writer.writerow(['sseq', 'qseq', 'start_pos'])
    for k, v in full_res.items():
        for a in v:
            promoter = a[0]
            for b in a[1:]:
                writer.writerow([k, promoter, b + 1])
```
